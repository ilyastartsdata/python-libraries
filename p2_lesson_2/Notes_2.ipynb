## <center>Theoretical part<a class="anchor" id="theory"></a><center>

### Probability theory and mathematical statistics<a class="anchor" id="prob_theory_and_math_stat">

**Probability theory** studies models of random variables and the properties of those models.

**Mathematical statistics and data analysis** tries to determine the properties of a random variable from the properties of finite samples in order to understand how it will behave in the future.

### What is a statistical hypothesis?<a class="anchor" id="stat_hypothesis"></a>

**A statistical hypothesis** is an assumption about the distribution and properties of a random variable that can be confirmed or disproved by applying statistical methods to sample data.

**The null hypothesis** is some assumption, accepted by default, that there is no relationship between the two observed events, deviation of indicators and other unexpected results, in a word there is no effect.

**Alternative hypothesis** - the alternative is usually the assumption being tested, but it also happens that the alternative is not explicitly stated, in which case the null hypothesis statement is rejected.

**Statistical hypothesis testing** is the process of deciding whether the statistical hypothesis in question is contrary to the observed sample of data.

**A statistical test or statistical criterion** is a strict mathematical rule by which a statistical hypothesis is accepted or rejected.

** An example of a formalised description of a hypothesis**

$\large x^n = (x_1, ... , x_n), \: x^n \in X, \: X \sim P$<br>
$\large H_0: \: P \in \omega$<br>
$\large H_1: \: P \notin \omega$<br>
$\large T(x^n), \: T(x^n) \sim F_0(t) \: | \: H_0, \: T(x^n) \nsim F_0(t) \: |\: H_1$<br>

$\normalsize H_0$ - null hypothesis<br>
$\normalsize H_1$ - alternative hypothesis<br>
$\normalsize X$ - random variable<br>
$\normalsize x^n$ - sample size $n$ of random variable $X$<br>
$\normalsize P$ - some distribution of random variable $X$<br>
$\normalsize \omega$ - some family of distributions<br>
$\normalsize T(x^n)$ - statistics from sample $x^n$<br>
$\normalsize F_0(t)$ - null distribution of the statistic<br>

This example tests the hypothesis $\normalsize H_0$ that the distribution $\normalsize P$ of a random variable $\normalsize X$, belongs to a family of distributions that we have defined in advance, assuming it is a family of normal distributions. Alternatively, the hypothesis $\normalsize H_1$ states that the distribution $\normalsize P$ belongs to some other family of distributions.

The statistic $\normalsize T(x^n)$ and its null distribution $\normalsize F_0(t)$ form the *statistical test*.

### Statistical hypothesis test<a class="anchor" id="stat_hypothesis_test"></a>

**Methodology for testing statistical hypotheses**.

1. Formulate hypotheses $\normalsize H_0$ and $\normalsize H_1$
2. Choose a suitable statistical criterion based on the hypothesis formulation, sample size(s), etc.
3. Record the level of significance $\normalsize \alpha$
4. On set of values of selected statistic $\normalsize T$ determine critical region $\normalsize \Omega_\alpha$ of least likely values, such that $\normalsize P(T \in \Omega_\alpha | H_0) = \alpha$, usually considered bilateral critical region: 
$\normalsize (-\infty;x_{\alpha/2}) \cup (x_{1 - \alpha/2};+\infty)$
5. Calculate statistic $\normalsize T$ and achievable significance level $\normalsize p-value^* = P(T \ge t \: | \: H_0)$
6. If $\normalsize p-value < \alpha$, $\normalsize H_0$ is rejected in favour of $\normalsize H_1$, as the probability of obtaining such data (sample), if $\normalsize H_0$ is true, is extremely low.

The *achievable significance level, p-value* is the probability, if the null hypothesis is true, to obtain the same distribution of statistics as in the experiment, or even more extreme.

<img src='images/p_value.png'>

**First- and second-order errors**

The first kind of error is when the null hypothesis is rejected although it is in fact true.<br>
The second kind of error is when the null hypothesis is accepted, although it is not actually true.<br>

|$H_0$ |**false** |**false** |
|:----------------:|:--------------------------------------------:|:-----------------------------------------:|
|**accepted** |$H_0$ faithfully accepted |$H_0$ incorrectly accepted (error of the second kind)|
|**rejected** |$H_0$ incorrectly rejected (error of the first kind)|$H_0$ correctly rejected |

In the mechanism of hypothesis testing, errors of the first and second kind are unequal; an error of the first kind is more critical, any correct statistical criterion must ensure that the probability of an error of the first kind is no greater than $\normalsize \alpha$,<br>
$\normalsize P(H_0 \: rejected \: | \: H_0) = P (p \le \alpha \: | \: H_0) \le \alpha$

The second kind of error is related to the notion of statistical power,<br>
$\normalsize pow = P(H_0 \: rejected \: | \: H_1) = 1 - P(H_0 \: accepted \: | \: H_1)$ is the probability of rejecting the null hypothesis when the alternative is true.

### Shapiro-Wilk criterion<a class="anchor" id="shapiro_criterion"></a>

This criterion tests the hypothesis that a random variable has a normal distribution (Gaussian distribution). The need to check a random variable for "normality" is due to the fact that many statistical criteria and analytical methods from mathematical statistics are focused on samples of normally distributed random variables and before using them you should make sure that the distribution law is close to normal.

In addition, normally distributed random variables have some properties that may be useful when dealing with them.

**Normal distribution**

<img src="images/normal_distr.png">

**Formalised description**

$\large x^n = (x_1, .. , x_n), \: x^n \in X$<br>

$\large H_0: \: X \sim N(\mu, \sigma^2)$<br>
$\large H_1: \: H_0 \: неверна$<br>

$\large W(x^n) = \frac{(\sum_{i=1}^n a_i x_i)^2}{\sum_{i=1}^n (x_i - \bar{x})^2} $<br>

The null distribution of statistics is tabular.

### Student's t-test, two-sample<a class="anchor" id="student_criterion"></a>

Student's test is a common name for statistical tests in which the criterion statistic has a Student's distribution.<br> Most often these criteria are used to check for equality of mean values (mat expectations) in two samples.

**Formalised description**

$\large x_1^{n_1} = (x_{11}, .. , x_{1{n_1}}), \: x_1^{n_1} \in X_1, \: X_1 \sim N(\mu_1, \sigma_1^2)$<br>
$\large x_2^{n_2} = (x_{21}, .. , x_{2{n_1}}), \: x_2^{n_2} \in X_2, \: X_2 \sim N(\mu_2, \sigma_2^2)$<br>

$\large H_0: \: \mu_1 = \mu_2$<br>
$\large H_1: \: \mu_1 \ne \mu_2$<br>

$\large T(x_1^{n_1}, x_2^{n_2}) = \frac{\bar{x_1} \: - \: \bar{x_2}}{\sqrt{\frac{S_1^2}{n_1} - \frac{S_2^2}{n_2}}}$<br>
$\large T(x_1^{n_1}, x_2^{n_2}) \sim St$

**Application conditions**

* Normal distribution, no outliers
* Sample size of at least 30 observations

If the data do not meet these criteria, *the Mann-Whitney U test is applied*, which is a non-parametric test that uses not the raw data, but their rank positions for calculation.

### Chi-square test (Pearson's goodness-of-fit test)<a class="anchor" id="pirson_criterion"></a>

The chi-square test assesses the significance of differences between the actual (revealed by the study) number of outcomes and the theoretical number that can be expected in the groups under study if the null hypothesis is valid. In simpler terms, the method allows us to assess the statistical significance of differences between two or more relative indicators (frequencies, fractions).

**Formalised description**

$\large x^n = (x_1, ... , x_n), \: x^n \in X$<br>

$\large H_0: \: Empirical \: (observed) \: and \: theoretical \: (expected) \: frequencies \: agree$ <br>
$\large H_1: \: H_0 \: incorrect<br>

$\large \chi^2(x^n) = \sum_{i=1}^K \frac{(O_i - E_i)^2}{E_i}$<br>
$O$ (Observed) - observed frequencies<br> 
$E$ (Expected) - expected frequencies<br>
$K$ - number of estimated frequencies

$\large \chi^2(x^n) \sim \chi^2$

**Application conditions**

The groups to be compared must be independent, i.e. chi-square test should not be applied when comparing before-after observations or related pairs. An analogue for dependent samples is the *McNemar test* or the *Q-test of Cochran* for comparing three or more groups.

If there are fewer than 10 observations in a cell, the *Yates correction* is applied. 

If less than 5, the *Fisher exact test* is used instead of the chi-square.

### Confidence intervals<a class="anchor" id="conf_interval"></a>

A type of interval estimate which sets numerical limits within which, with a certain probability, the true value of the parameter being estimated is found.

**The procedure for calculating the confidence interval (for the mathematical expectation)**

1. Define confidence level, $\normalsize \alpha = 95\% = 0.95$
2. Refer to Z-score table or calculate confidence coefficient - $\normalsize Z_{\alpha/2}$, for $\normalsize \alpha = 0.95, Z_{\alpha/2} = 1.96$
3. Calculate confidence interval, $\normalsize CI = \bar{x} \pm Z_{\alpha/2} \frac{\sigma}{\sqrt{n}}$,<br>
where $\normalsize \bar{x}$ is the sample mean, $\normalsize \sigma$ is the standard deviation, $\normalsize n$ is the sample size

**Identification of target variable and feature groups**

```python
TARGET_NAME = 'NEXT_MONTH_DEFAULT'
BASE_FEATURE_NAMES = df_base.columns.drop(TARGET_NAME).tolist()
NEW_FEATURE_NAMES = df.columns.drop([TARGET_NAME] + BASE_FEATURE_NAMES)

NUM_FEATURE_NAMES = ['LIMIT_BAL', 'AGE',
                     'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6',
                     'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']

CAT_FEATURE_NAMES = ['SEX', 'EDUCATION', 'MARRIAGE', 'PAY_1', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
```

**Correlation with basic attributes**

```python
corr_with_target = df[BASE_FEATURE_NAMES + [TARGET_NAME]].corr().iloc[:-1, -1].sort_values(ascending = False)

plt.figure(figsize = (10, 8))

sns.barplot(x = corr_with_target.values, y = corr_with_target.index)

plt.title('Correlation with target variable')
plt.show()
```

**Assessment of the "PAY_1" attribute by target variable**

```python
plt.figure(figsize = (10, 8))

sns.countplot(x = "PAY_1", hue = TARGET_NAME, data = df)
plt.title('PAY_1 grouped by target variable')
plt.legend(title = 'Target', loc = 'upper right')

plt.show()
```

**Observation**.

Examining the resulting graph, we can see that the values -1 (redeemed completely) and 0 (redeemed partially) of the attribute PAY_1 have similar proportions in terms of the target variable. If this is indeed the case, it would be possible to combine them into a single category, for example.

**Hypothesis**. 

* Null hypothesis: expected and observed frequencies are in agreement
* Alternative hypothesis: deviations in frequencies are beyond random fluctuations, discrepancies are statistically significant
* Criterion: Pearson Chi-Square
* Significance level $\normalsize \alpha$: 0.05
* Critical area: two-sided

To test this hypothesis, you need to feed the observed frequencies of categories -1 and 0 of the attribute PAY_1 into the selected criterion, then assess the value of the achieved significance level p-value and compare it to the selected threshold alpha, if p-value turns out to be greater than the selected threshold, then the hypothesis of consistency of frequencies can not be discarded.

*Calculate the observed frequencies*

```python
df['Id'] = df.index
pay1_and_target_s = df.loc[df['PAY_1'].isin([-1, 0]), ['Id', 'PAY_1', 'NEXT_MONTH_DEFAULT']]
```

```python
table = pay1_and_target_s.pivot_table(values='Id', index='PAY_1', columns='NEXT_MONTH_DEFAULT', aggfunc='count')
table
```

*Test our hypothesis using the Pearson Chi-square test*

```python
chi2, p, _, _ = chi2_contingency(table)
p
```

```python
pay1_and_target = df.loc[df['PAY_1'].isin([2, 3]), ['Id', 'PAY_1', 'NEXT_MONTH_DEFAULT']]
table = pay1_and_target.pivot_table(values = 'Id', index = 'PAY_1', columns = 'NEXT_MONTH_DEFAULT', aggfunc = 'count')

chi2, p, _, _ = chi2_contingency(table)
p
```

The P-value is greater than the chosen level of significance, so there is no reason to reject the null hypothesis and we can assume that categories 2 and 3 have the same effect on the target variable and they can be combined into one category. 

\*For the other attributes PAY_2, PAY_3, etc. a similar analysis should be conducted and then we should decide whether to change the categories or construct new attributes.

**Assessment of the "LIMIT_BAL" attribute in terms of the target variable**

```python
limit_bal_with_target = df[['LIMIT_BAL', TARGET_NAME]]
limit_bal = limit_bal_with_target['LIMIT_BAL']
limit_bal_target_0 = limit_bal[limit_bal_with_target[TARGET_NAME] == 0]
limit_bal_target_1 = limit_bal[limit_bal_with_target[TARGET_NAME] == 1]

plt.figure(figsize = (10, 5))

sns.kdeplot(limit_bal_target_0, shade=True, label = 'Нет просрочки', color = 'g')
sns.kdeplot(limit_bal_target_1, shade=True, label = 'Просрочка', color = 'r')

plt.xlabel('LIMIT_BAL')
plt.title('LIMIT_BAL grouped by target variable')
plt.show()
```

**Observation**.

It appears that the two groups resulting from partitioning the "LIMIT_BAL" trait by target variable have different distributions, which may help in constructing the model as it would imply that there may be some functional relationship between the "LIMIT_BAL" trait and the target variable.

**Hypothesis**. 
*Null hypothesis: the mean values in two independent samples are equal
* Alternative hypothesis: two independent samples have different mean values
* Criterion: Student's t-test and its analogues
* Significance level $\normalsize \alpha$: 0.05
* Critical area: two-sided

To test this hypothesis, compare two samples from the groups under consideration for equality of mean values.
If the probability of the mathematical expectation being equal in the original groups is less than 5% for the given samples, then we can say that it is likely that the samples have different distributions.

*Checking the distribution for "normality" using the Shapiro-Wilk criterion*

```python
shapiro(limit_bal)
```

From the resulting p-value, which is much lower than 0.05, we can conclude that the hypothesis of "normality" is rejected.

*For plausibility, we make a visual assessment of the distribution of the trait, and construct a QQ plot*

```python
plt.figure(figsize = (20, 8))

ax1 = plt.subplot(121)
ax1.set_xlabel('LIMIT_BAL')
ax1.set_ylabel('Count')
ax1.set_title('LIMIT_BAL distribution')
limit_bal.hist()

plt.subplot(122)
probplot(limit_bal, dist = 'norm', plot = plt)

plt.show()
```

The visual assessment confirms the Shapiro-Wilk test indicating that the distribution law is different from "normal", therefore, we cannot use Student's test to test the hypothesis of equal expectation of LIMIT_BAL in the groups with overdue and not overdue payments in the next month, but we can use its nonparametric counterpart, the Mann-Whitney test, which is not subject to the distribution law.

*Assess the equivalence of the matrices, in the study groups, using the Mann-Whitney test*

```python
mannwhitneyu(limit_bal_target_0, limit_bal_target_1)
```

According to the p-value, the hypothesis of equal expectations is rejected, but it is worth referring to the confidence intervals for additional verification.

*Line up the confidence intervals for the mean values of each of the two groups and compare them*

```python
plt.figure(figsize = (8, 5))

sns.pointplot(x = TARGET_NAME, y = 'LIMIT_BAL', data = limit_bal_with_target, capsize = .1)

plt.title('Confidence intervals (95 %) for LIMIT_BAL')
plt.show()
```

This graph also shows that the intervals in which the true expectations of the two groups should be found with 95% probability do not overlap, which confirms the Mann-Whitney test results.

This means that the groups from which the sample data are taken have different distributions with 95% probability and this feature can be useful for determining the value of the target variable.
